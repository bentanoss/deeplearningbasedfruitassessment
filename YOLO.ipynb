```python
from ultralytics import YOLO
import os, time, csv
from datetime import datetime

# 📁 Final run setup
SAVE_DIR = "/content/drive/MyDrive/yolo_results_final_run3"
os.makedirs(SAVE_DIR, exist_ok=True)

results_file = os.path.join(SAVE_DIR, "model_variant_final_comparison.csv")

# 📄 Create CSV for logging (only create header if the file doesn't exist)
if not os.path.exists(results_file):
    with open(results_file, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([
            "Timestamp", "Model",
            "Train_Precision", "Train_Recall", "Train_mAP50", "Train_F1",
            "Test_Precision", "Test_Recall", "Test_mAP50", "Test_F1",
            "Training Time (s)", "Model Path"
        ])

# 🧪 Model variants
model_variants = ["yolov8m.pt"]

# 🚀 Train and Evaluate
for model_variant in model_variants:
    model_name = model_variant.split(".")[0]
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    run_name = f"{model_name}_final_{timestamp}"
    model_path = os.path.join(SAVE_DIR, f"{run_name}_best.pt")

    print(f"\n🚀 Training: {model_variant}")
    model = YOLO(model_variant)

    start_time = time.time()
    model.train(
        data="/content/fresh-and-rotten-fruits-2/data.yaml",
        epochs=50,
        imgsz=640,
        batch=16,
        lr0=0.001,
        weight_decay=0.0001,
        optimizer="AdamW",
        project=SAVE_DIR,
        name=run_name,
        save=True,
        plots=True,
    )
    training_time = time.time() - start_time

    # 💾 Save trained model
    best_model = f"runs/detect/{run_name}/weights/best.pt"
    os.system(f"cp {best_model} {model_path}")

    # 📊 Evaluate on validation set
    print(f"📈 Evaluating {model_name} on validation set...")
    val_metrics = model.val(data="/content/fresh-and-rotten-fruits-2/data.yaml")

    val_precision = val_metrics.box.mp
    val_recall = val_metrics.box.mr
    val_map50 = val_metrics.box.map50
    val_f1 = 2 * (val_precision * val_recall) / (val_precision + val_recall + 1e-6)

    # 📊 Evaluate on test set
    print(f"🧪 Evaluating {model_name} on test set...")
    test_metrics = model.val(data="/content/fresh-and-rotten-fruits-2/data.yaml", split='test')

    test_precision = test_metrics.box.mp
    test_recall = test_metrics.box.mr
    test_map50 = test_metrics.box.map50
    test_f1 = 2 * (test_precision * test_recall) / (test_precision + test_recall + 1e-6)

    # 📝 Save to CSV
    with open(results_file, mode='a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([
            datetime.now(), model_name,
            round(val_precision, 4), round(val_recall, 4), round(val_map50, 4), round(val_f1, 4),
            round(test_precision, 4), round(test_recall, 4), round(test_map50, 4), round(test_f1, 4),
            round(training_time, 2), model_path
        ])
```
